## 5. How do I apply multiple filter criteria to a pandas DataFrame?

```python
# read a dataset of top-rated IMDb movies into a DataFrame
df1 = pd.read_csv('http://bit.ly/imdbratings')
df1.head()
```

```python
# filter the DataFrame to only show movies with a 'duration' of at least 200 minutes
df1[df1.duration >= 200]
```

```python
# CORRECT: use the '&' operator to specify that both conditions are required
df1[(df1.duration >=200) & (df1.genre == 'Drama')]
```

```python
# INCORRECT: using the '|' operator would have shown movies that are either long or dramas (or both)
df1[(df1.duration >=200) | (df1.genre == 'Drama')].head()
```

```python
# use the '|' operator to specify that a row can match any of the three criteria
df1[(df1.genre == 'Crime') | (df1.genre == 'Drama') | (df1.genre == 'Action')].head(10)

# or equivalently, use the 'isin' method
df1[df1.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```

## 6. When should I use the "inplace" parameter in pandas?

```python
# read a dataset of UFO reports into a DataFrame
df = pd.read_csv('http://bit.ly/uforeports')
df.head()
```

```python
df.shape
```

```python
# remove the 'City' column (doesn't affect the DataFrame since inplace=False)
df.drop('City', axis=1).head()
```

```python
# confirm that the 'City' column was not actually removed
df.head()
```

```python
# remove the 'City' column (does affect the DataFrame since inplace=True)
df.drop('City', axis=1, inplace=True)
```

```python
# confirm that the 'City' column was actually removed
df.head()
```

```python
# drop a row if any value is missing from that row (doesn't affect the DataFrame since inplace=False)
df.dropna(how='any').shape
```

```python
# confirm that no rows were actually removed
df.shape
```

```python
# use an assignment statement instead of the 'inplace' parameter
df = df.set_index('Time')
df.tail()
```

```python
# fill missing values using "backward fill" strategy (doesn't affect the DataFrame since inplace=False)
df.fillna(method='bfill').tail()
```

```python
# compare with "forward fill" strategy (doesn't affect the DataFrame since inplace=False)
df.fillna(method='ffill').tail()
```

## 1. When should I use a "groupby" in pandas?

```python
# read a dataset of UFO reports into a DataFrame
import pandas as pd
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
# calculate the mean beer servings across the entire dataset
drinks.beer_servings.mean()
```

```python
# calculate the mean beer servings just for countries in Africa
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
# calculate the mean beer servings for each continent
drinks.groupby('continent').beer_servings.mean()
```

```python
# other aggregation functions (such as 'max') can also be used with groupby
drinks.groupby('continent').beer_servings.max()
```

```python
# multiple aggregation functions can be applied simultaneously
drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
# specifying a column to which the aggregation function should be applied is not required
drinks.groupby('continent').mean()
```

```python
import matplotlib.pyplot as plt

# allow plots to appear in the notebook
%matplotlib inline
```

```python
# side-by-side bar plot of the DataFrame directly above
drinks.groupby('continent').mean().plot(kind='bar')
```

## 2. How do I handle missing values in pandas?- Simple imputer and types of imputation

```python
# read a dataset of UFO reports into a DataFrame
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.tail()
```

```python
# 'isnull' returns a DataFrame of booleans (True if missing, False if not missing)
ufo.isnull().tail()
```

```python
# 'nonnull' returns the opposite of 'isnull' (True if not missing, False if missing)
ufo.notnull().tail()
```

```python
# count the number of missing values in each Series
ufo.isnull().sum()
```

```python
# use the 'isnull' Series method to filter the DataFrame rows
ufo[ufo.City.isnull()].head()
```

```python
# examine the number of rows and columns
ufo.shape
```

```python
# if 'any' values are missing in a row, then drop that row
ufo.dropna(how='any').shape
```

```python
# 'inplace' parameter for 'dropna' is False by default, thus rows were only dropped temporarily
ufo.shape
```

```python
# if 'all' values are missing in a row, then drop that row (none are dropped in this case)
ufo.dropna(how='all').shape
```

```python
# if 'any' values are missing in a row (considering only 'City' and 'Shape Reported'), then drop that row
ufo.dropna(subset=['City', 'Shape Reported'], how='any').shape
```

```python
# if 'all' values are missing in a row (considering only 'City' and 'Shape Reported'), then drop that row
ufo.dropna(subset=['City', 'Shape Reported'], how='all').shape
```

```python
# 'value_counts' does not include missing values by default
ufo['Shape Reported'].value_counts().head()
```

```python
# explicitly include missing values
ufo['Shape Reported'].value_counts(dropna=False).head()
```

```python
# fill in missing values with a specified value
ufo['Shape Reported'].fillna(value='VARIOUS', inplace=True)
```

```python
# confirm that the missing values were filled in
ufo['Shape Reported'].value_counts().head()
```

### Four options for handling missing values (NaNs):

Drop rows containing NaNs

Drop columns containing NaNs

Fill NaNs with imputed values

Use a model that natively handles NaNs (NEW!)

```python
import pandas as pd
train = pd.read_csv('http://bit.ly/kaggletrain')
test = pd.read_csv('http://bit.ly/kaggletest', nrows=175)

train = train[['Survived', 'Age', 'Fare', 'Pclass']]
test = test[['Age', 'Fare', 'Pclass']]

# count the number of NaNs in each column
train.isna().sum()
```

```python
test.isna().sum()
```

```python
label = train.pop('Survived')
```

```python
# new in 0.22: this estimator (experimental) has native support for NaNs
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
```

```python
clf = HistGradientBoostingClassifier()
```

```python
# no errors, despite NaNs in train and test!
clf.fit(train, label)
clf.predict(test)
```

### Need to impute missing values for a categorical feature?

Two options:

Impute the most frequent value

Impute the value "missing", which treats it as a separate category

```python
import numpy as np
X = pd.DataFrame({'Shape':['square', 'square', 'oval', 'circle', np.nan]})
X
```

## 1. How do I make my pandas DataFrame smaller and faster?

```python
import pandas as pd

data = pd.read_csv('http://bit.ly/drinksbycountry')
data.head()
```

```python
data.info()
```

### You can see that the memory usage above is 9.2+ KB. The question is how much + this is? Let's find out.

```python
data.info(memory_usage = 'deep')
```

### So the total memory is 30.5 KB. We can also check how much memory each column is taking in Bytes.

```python
data.memory_usage()
```

### There is one problem though. The column country and continent are object datatypes and object datatypes are more like pointers. It means they are not displaying the correct memory usage.

```python
data.memory_usage(deep = True)
```

### Now you can see the correct memory usage. We can verify the sum of these as well.

```python
data.memory_usage(deep = True).sum()
```

### How to optimize memory now? Let's start with the continent column first. We need to check how many unique values are there.

```python
sorted(data.continent.unique())
```

```python
data.continent.head()
```

### What if we can convert this object column into integer column? We can save a lot of memory and make our dataframe faster and smaller. We are talking about creating some kind of look up table here. It is done by converting the object data into category datatype.

```python
data['continent'] = data.continent.astype('category')
```

```python
data.dtypes
```

```python
data.continent.head()
```

### In the backend, category has created numeric codes to represent the continent column values. We can check these codes.

```python
data.continent.cat.codes.head()
```

### Let's verify if this has something to do with our memory reduction part.

```python
data.memory_usage(deep = True)
```

### As you can see that the continent memory which was 12000 + Bytes has now reduced to 756 Bytes. So we have made our dataframe smaller and faster. Now let's try and do the same on country column.

```python
data['country'] = data.country.astype('category')
```

```python
data.dtypes
```

```python
data.memory_usage(deep = True)
```

### OOPS!! Instead of reducing the memory, the memory size is increased. What happened here? Let's check the unique values present in the country column.

```python
data.country.unique()
```

```python
data.continent.unique()
```

### You can see that there are 6 unique values in continent column which are used 193 times. Category is encoding these 6 unique values. But there are 193 unique values in country column which are used once 193 times. So all values are unique. The concept is that use category datatype when there is less number of unique values. Otherwise memory size will increase.

## 2. How do I create dummy variables in pandas?

```python
import pandas as pd

data = pd.read_csv('http://bit.ly/kaggletrain')
data.head()
```

### There are 2 categorical columns: Sex and Embarked. Sex column has 2 categories- Male and Female. Embarked column has 3 categories- S, C and Q. Dummy values are used to convert categorical values to numeric values because machine don't understand the categorical values. It understands only the numbers.

```python
pd.get_dummies(data.Sex)
```

### We can remove any one column. The remaining column will be enough to analyse the value of other column as well. For example, 1 means Male and 0 means Not Male.

```python
pd.get_dummies(data.Sex, prefix = 'sex').iloc[:, 1:]
```

### Similarly, we can check for Embarked column as well. First, let's check the values present in this column.

```python
data.Embarked.value_counts()
```

```python
pd.get_dummies(data.Embarked, prefix = 'Embarked')
```

### We can remove any one column. The remaining two columns will be enough to determine the value of 3rd column as well. 

```python
pd.get_dummies(data.Embarked, prefix = 'Embarked').iloc[:, 1:]
```

### Let's make this change permanent in the dataset as well.

```python
embarked_dummies = pd.get_dummies(data.Embarked, prefix = 'Embarked').iloc[:, 1:]
gender_male = pd.get_dummies(data.Sex, prefix = 'sex').iloc[:, 1:]
data = pd.concat([data, gender_male, embarked_dummies], axis = 1)
data.head()
```

```python
data.drop(['Sex', 'Embarked'], axis = 1, inplace = True)
```

```python
data.head()
```
